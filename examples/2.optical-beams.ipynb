# Utilização básica do construtor de feixes

```python
# módulo de controle simples de arrays
import numpy as np
# módulo de produção de gráficos
import matplotlib.pyplot as plt

# construção da janela computacional
l = np.linspace(-40.0, +40.0, 500); # lista de pontos da partição dos eixos x e y
X, Y = np.meshgrid(l,l);            # matriz de x e y dos pontos da malha de partição

# computação dos limites dos eixos
extremos = [
    l[0], l[-1],                    # extremos do eixo x
    l[0], l[-1]                     # extremos do eixo y
];
```

Ademais à criação de campos ópticos genéricos podemos construir diretamente feixes ópticos pelos construtores do módulo `beam`.

```python
# retornar ao diretório raíz do projeto
%cd -q ../
# importa o módulo beam do módulo optical
from opticalField import beam
# importa as funções intensity e phase do módulo opticalField
from opticalField import intensity, phase
```

## Construção de feixe gaussiano por `beam.Gaussian`

A construção de um feixe gaussiano foi implementada pelo construtor `beam.Gaussian()` onde a envoltória de onda não é passada como argumento (sendo substituida pelo parâmetro `w0` que corresponde à cintura do feixe).

```python
gaussianBeam = beam.Gaussian(
    w0 = 20.0,
    region = (X, Y)
);
```

### Perfil de intensidade transversal do feixe gaussiano

```python
# criação de figuras para plotagem de imagem
fig, ax = plt.subplots();

# rotulação do título da figura
ax.set_title('Intensidade');

# rotulação dos eixos coordenados
ax.set_xlabel('x($\mu$m)')
ax.set_ylabel('y($\mu$m)')

# geração de imagem do perfil de intesidade do feixe gaussiano
imageIntensity = ax.imshow(
    intensity(gaussianBeam),
    cmap = 'inferno',               # define o mapa de cores
    extent = extremos               # define a extensão dos eixos x e y
);
fig.colorbar(imageIntensity);       # insere uma colorbar
```

### Perfil de fase transversal do feixe gaussiano

```python
# criação de figuras para plotagem de imagem
fig, ax = plt.subplots();

# rotulação do título da figura
ax.set_title('Fase');

# rotulação dos eixos coordenados
ax.set_xlabel('x($\mu$m)')
ax.set_ylabel('y($\mu$m)')

# geração de imagem do perfil de fase do feixe gaussiano
imageIntensity = ax.imshow(
    phase(gaussianBeam),
    cmap = 'cividis',               # define o mapa de cores
    extent = extremos               # define a extensão dos eixos x e y
);
fig.colorbar(imageIntensity);       # insere uma colorbar
```

## Construção de feixe Hermite-Gauss por `beam.HermiteGauss`

A construção de um feixe hermite-gauss de índices $(l,m)$ foi implementada pelo construtor `beam.HermiteGauss()` onde a envoltória de onda não é passada como argumento (sendo substituida pelo parâmetro `w0` que corresponde à cintura do feixe a `2-tupla` de índices `index`).

```python
hermitegaussBeam = beam.HermiteGauss(
    index = (0, 2),
    w0 = 20.0,
    region = (X, Y)
);
```

### Perfil de intesidade transversal do feixe hermite-gauss

```python
# criação de figuras para plotagem de imagem
fig, ax = plt.subplots();

# rotulação do título da figura
ax.set_title('Intensidade');

# rotulação dos eixos coordenados
ax.set_xlabel('x($\mu$m)')
ax.set_ylabel('y($\mu$m)')

# geração de imagem do perfil de intesidade do feixe hermite-gauss
imageIntensity = ax.imshow(
    intensity(hermitegaussBeam),
    cmap = 'inferno',               # define o mapa de cores
    extent = extremos               # define a extensão dos eixos x e y
);
fig.colorbar(imageIntensity);       # insere uma colorbar
```

### Perfil de fase transversal do feixe hermite-gauss

```python
# criação de figuras para plotagem de imagem
fig, ax = plt.subplots();

# rotulação do título da figura
ax.set_title('Fase');

# rotulação dos eixos coordenados
ax.set_xlabel('x($\mu$m)')
ax.set_ylabel('y($\mu$m)')

# geração de imagem do perfil de fase do feixe hermite-gauss
imageIntensity = ax.imshow(
    phase(hermitegaussBeam),
    cmap = 'cividis',               # define o mapa de cores
    extent = extremos               # define a extensão dos eixos x e y
);
fig.colorbar(imageIntensity);       # insere uma colorbar
```
